```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [43, 39, 42.5, 41, 40, 44, 45]
min_t = [23, 21, 25, 21, 21, 19, 26]
avg_t = [34, 35, 35, 35, 30, 29, 33]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temp.(C)')
plt.title('Weather')
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temp.(C)')
plt.title('Weather')
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Temp.(C)')
plt.title('Weather')
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend(loc='upper right', shadow=True, fontsize='small')
```
