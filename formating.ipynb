```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3,4 ,5,6,7]
y = [12, 34, 2, 12, 21, 9, 46]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'g+')
```

```python
plt.plot(x, y, 'g+--')
```

```python
plt.plot(x, y, 'gD-.')
```

```python
plt.plot(x, y, color = 'red', marker='D', linestyle='-.')
```

```python
plt.plot(x, y, color = 'red', marker='D', linestyle='-.', alpha=0.5)
```
