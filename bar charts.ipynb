```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import numpy as np
```

```python
comp = ['gogl', 'amzn', 'msft', 'apl', 'fb']
rev = [120, 90, 75, 132, 83]
```

```python
xpos = np.arange(len(comp))
xpos
```

```python
plt.bar(xpos, rev)
```

```python
plt.xticks(xpos, comp)
plt.bar(xpos, rev)
```

```python
plt.xticks(xpos, comp)
plt.xlabel('Companies')
plt.ylabel('Revenues')
plt.title('Revenues of Companies')
plt.bar(xpos, rev, label='revenue')
plt.legend()
```

```python
profit = [75, 54, 23, 100, 41]
```

```python
plt.xticks(xpos, comp)
plt.xlabel('Companies')
plt.ylabel('Revenues')
plt.title('Revenues of Companies')
plt.bar(xpos, rev, label='revenue')
plt.bar(xpos, profit, label='profits')
plt.legend()
```

```python
plt.xticks(xpos, comp)
plt.xlabel('Companies')
plt.ylabel('Revenues')
plt.title('Revenues of Companies')
plt.bar(xpos-0.2, rev, label='revenue')
plt.bar(xpos+0.2, profit, label='profits')
plt.legend()
```

```python
plt.xticks(xpos, comp)
plt.xlabel('Companies')
plt.ylabel('Revenues')
plt.title('Revenues of Companies')
plt.bar(xpos-0.2, rev, width=0.4, label='revenue', color='#c4abfc')
plt.bar(xpos+0.2, profit, width=0.4, label='profits',color='#ded1fc')
plt.legend()
plt.savefig("bar.pdf", bbox_inches='tight',pad_inches=1)
```

```python
plt.yticks(xpos, comp)
plt.ylabel('Companies')
plt.xlabel('Revenues')
plt.title('Revenues of Companies')
plt.barh(xpos-0.2, rev, label='revenue')
plt.barh(xpos+0.2, profit, label='profits')
plt.legend()
```

```python
plt.yticks(xpos, comp)
plt.ylabel('Companies')
plt.xlabel('Revenues')
plt.title('Revenues of Companies')
plt.barh(xpos-0.2, rev, height= 0.4, label='revenue')
plt.barh(xpos+0.2, profit, height = 0.4, label='profits')
plt.legend()
```
